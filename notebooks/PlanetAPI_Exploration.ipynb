```python
image_id = '20240514_172122_81_2470'
```

```python
assets_url = f"https://api.planet.com/data/v1/item-types/PSScene/items/{image_id}/assets/"
```

```python
response = requests.get(assets_url, headers=headers)
assets = response.json()
```

```python
asset_type = "ortho_visual" # We can explore other types (infrared etc. when needed)
```

```python
activate_url = assets[asset_type]['_links']['activate']
requests.get(activate_url, headers=headers)
print(f"Activating {asset_type}, wait time starts...") # Image is not readily available. It needs some additional processing. Hence the wait time, and the term "activate".
```

```python
while True:
    response = requests.get(assets_url, headers=headers)
    assets = response.json()
    asset_status = assets[asset_type]['status']

    if asset_status == 'active':
        download_url = assets[asset_type]['location']

        image_data = requests.get(download_url, headers=headers)

        with open(f"{image_id}.tif", "wb") as f:
            f.write(image_data.content)

        print(f"Image {image_id} downloaded successfully!")
        break
    else:
        # Wait for 30 seconds per cycle
        print(f"Asset is still {asset_status}, checking again in 30 seconds...")
        time.sleep(30)
```

```python
image_path = f"{image_id}.tif"
image = Image.open(image_path)

plt.figure(figsize=(8, 8))
plt.imshow(image)
plt.axis('off')
plt.show()
```
